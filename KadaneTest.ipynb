```
def larger0sub(testarray):
    """ 
    Esta función retorna el índice izquierdo y la longitud del sub arreglo con 
    la mayor cantidad de ceros contiguos del arreglo <<testarray>> 
 
    Input: 
    testarray (numpy array): 1D numpy array con ceros (0) y unos (1). 
  
    Salida: 
    ilbest,len(best_array) (tuple): tupla con los enteros correspondientes al 
    índice desde donde empieza el sub arreglo y a la longitud (número de elemen-
    tos) del mismo.  
    """ 
    zeroes,=np.where(testarray==0)#Busca los índices con ceros en testarray. 
    c=0#Contador
    ilbest=zeroes[0]#Inicialización
    irbest=zeroes[0]#
    best_array=testarray[zeroes[0]:zeroes[0]]
    while (c<(len(zeroes))):
        ileft=zeroes[c]
        if (ileft+1)<len(testarray):   
            iright=ileft+1
        else:
            iright=ileft
        while (testarray[iright]==0):#Expande el subarreglo hasta encontrar un 1
            iright=iright+1
            c=c+1
            if (iright)==len(testarray):#Rompe el ciclo cuando llega al extremo
                iright=iright
                break 
        c=c+1
        cur_array=testarray[ileft:iright]
        if (len(cur_array)>len(best_array)):#Actualiza el subarreglo más grande 
            best_array=cur_array
            ilbest=ileft
            irbest=iright
    return ilbest,len(best_array)
```

```
def maximo_rectangulo(A):   
    """
    Esta función retorna la fila y la columna (de la esquina superior izquierda)
    y el ancho de la submatriz cuadrada con mayor cantidad de ceros contiguos de
    la matriz A
    """
    Aarray=np.array(A)
    nrows,ncols=Aarray.shape
    Area=1
    for left in range(ncols):
        for right in range(left+1,ncols+1):
            if (0 in Aarray[:,left:right].sum(1)):
                Up,height=larger0sub(Aarray[:,left:right].sum(1))
                if (height*(right-left)>Area):
                    larger_rectang=(Up,left,right-left,height)
                    Area=(right-left)*height
    return larger_rectang
```

```
def maximo_cuadrado(A):
    """
    Esta función retorna la fila y la columna (de la esquina superior izquierda),
    el ancho y el largo de la submatriz rectangular con mayor cantidad de ceros 
    contiguos de la matriz A
    """
    Aarray=np.array(A)
    nrows,ncols=Aarray.shape
    Area=1
    for left in range(ncols):
        for right in range(left+1,ncols+1):
            if (0 in Aarray[:,left:right].sum(1)):
                Up,height=larger0sub(Aarray[:,left:right].sum(1))
                if ((height==(right-left)) & (height*(right-left)>Area)):
                    larger_square=(Up,left,height)
                    Area=(right-left)*height
    return larger_square
```

```
def isfree(listoftuples,Marray):
    csum=0
    for neighboor in listoftuples:
        csum=csum+Marray[neighboor[0],neighboor[1]]
    return csum==0
```
